# Only Back Sensor (raw)

## Training

```python
X, y, subjects = load_dataset_gait(base_path, process="raw", sensors=["LB"], window_size=200, stride=50)
X = X.astype(np.float32)
y = y.astype(np.int64)
```

```python
models_per_subject, results_list = train_model_cv(X, y, subjects, model_name="cnnbilstm", epochs=100, lr=0.0003, patience=10, out_dir="../checkpoints/CNNBiLSTM/raw_only_back")
```

## Analysis

```python
f1_scores = np.vstack((np.unique(subjects), results_list))
f1_scores = pd.DataFrame(f1_scores.T, columns=['patient', 'score'])
f1_scores.head()
```

```python
f1_scores.to_csv("../results/results_raw_LB_gait_cnnbilstm.csv", index=False)
```

```python
results_raw_LB = pd.read_csv("../results/results_raw_LB_gait_cnnbilstm.csv")
```

```python
print(f'Mean f1-score: {np.mean(results_raw_LB["score"])}')
print(f'Standard deviation: {np.std(results_raw_LB["score"])}')
```

```python
healthy = 'lightgreen'
ortho = 'lightcoral'
neuro = 'lightblue'

cohorts = {
    'HS': (("healthy", healthy), []),
    'ACL': (("ortho", ortho), []),
    'HOA': (("ortho", ortho), []),
    'KOA': (("ortho", ortho), []),
    'CIPN': (("neuro", neuro), []),
    'CVA': (("neuro", neuro), []),
    'PD': (("neuro", neuro), []),
    'RIL': (("neuro", neuro), [])}


for cohort in cohorts.keys():
    for patient_name in results_raw_LB['patient']:
        if cohort in patient_name:
            score = results_raw_LB.loc[results_raw_LB['patient'] == patient_name, 'score'].values[0]
            cohorts[cohort][1].append(score)

mean = np.mean(results_raw_LB["score"])

plt.figure(figsize=(12,8))

for cohort in cohorts.keys():
    bar = plt.bar(
        cohort,
        np.mean(cohorts[cohort][1]),
        yerr=np.std(cohorts[cohort][1]),
        capsize=5,
        label=cohorts[cohort][0][0],
        color=cohorts[cohort][0][1])

    plt.bar_label(bar, fmt='%.4f')

plt.axhline(y=mean, color='black', linestyle='--', label=f'Mean = {mean:.4f}')

plt.ylim(0.4, 1.05)

plt.legend()
plt.show()



```

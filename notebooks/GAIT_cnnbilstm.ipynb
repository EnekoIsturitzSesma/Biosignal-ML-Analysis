```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import gc 
from sklearn.metrics import f1_score

import sys
import os

sys.path.append(os.path.abspath(os.path.join('..')))

from src.load_data_gait import *
from src.plot_data_gait import *
from src.train_LSTMGait import train_model_cv, predict_trial
```

```python
base_path = '../datasets/GAIT_signals/dataset/data'
```

# Only Back Sensor (raw)

## Training

```python
X, y, subjects = load_dataset_gait(base_path, process="raw", sensors=["LB"], window_size=200, stride=50)
X = X.astype(np.float32)
y = y.astype(np.int64)
```

```python
models_per_subject, results_list = train_model_cv(X, y, subjects, model_name="cnnbilstm", epochs=100, lr=0.0003, patience=10, out_dir="../checkpoints/CNNBiLSTM/raw_only_back")
```

## Analysis

```python
f1_scores = np.vstack((np.unique(subjects), results_list))
f1_scores = pd.DataFrame(f1_scores.T, columns=['patient', 'score'])
f1_scores.head()
```

```python
f1_scores.to_csv("../results/results_raw_LB_gait_cnnbilstm.csv", index=False)
```

```python
results_raw_LB = pd.read_csv("../results/results_raw_LB_gait_cnnbilstm.csv")
```

```python
print(f'Mean f1-score: {np.mean(results_raw_LB["score"])}')
print(f'Standard deviation: {np.std(results_raw_LB["score"])}')
```

```python
healthy = 'lightgreen'
ortho = 'lightcoral'
neuro = 'lightblue'

cohorts = {
    'HS': (("healthy", healthy), []),
    'ACL': (("ortho", ortho), []),
    'HOA': (("ortho", ortho), []),
    'KOA': (("ortho", ortho), []),
    'CIPN': (("neuro", neuro), []),
    'CVA': (("neuro", neuro), []),
    'PD': (("neuro", neuro), []),
    'RIL': (("neuro", neuro), [])}


for cohort in cohorts.keys():
    for patient_name in results_raw_LB['patient']:
        if cohort in patient_name:
            score = results_raw_LB.loc[results_raw_LB['patient'] == patient_name, 'score'].values[0]
            cohorts[cohort][1].append(score)

mean = np.mean(results_raw_LB["score"])

plt.figure(figsize=(12,8))

for cohort in cohorts.keys():
    bar = plt.bar(
        cohort,
        np.mean(cohorts[cohort][1]),
        yerr=np.std(cohorts[cohort][1]),
        capsize=5,
        label=cohorts[cohort][0][0],
        color=cohorts[cohort][0][1])

    plt.bar_label(bar, fmt='%.4f')

plt.axhline(y=mean, color='black', linestyle='--', label=f'Mean = {mean:.4f}')

plt.ylim(0.4, 1.05)

plt.legend()
plt.show()



```

# Prediction Visualization

```python
best_subj_idx = np.argmax(results_raw_LB['score'])
best_subj = results_raw_LB.iloc[best_subj_idx]['patient'] 
print('Subject with the best score:')
print(results_raw_LB.iloc[best_subj_idx])
print('')

worst_subj_idx = np.argmin(results_raw_LB['score'])
worst_subj = results_raw_LB.iloc[worst_subj_idx]['patient'] 
print('Subject with the worst score:')
print(results_raw_LB.iloc[worst_subj_idx])
print('')
```

```python
best_subj_trial = f"{best_subj}_1"  # Obtain first trial of the best classified subject 
worst_subj_trial = f"{worst_subj}_1"  # Obtain first trial of the worst classified subject 

print(best_subj_trial)
print(worst_subj_trial)
```

```python
y_pred, y_true, signal = predict_trial(
    base_path   = base_path,
    trial_name  = best_subj_trial,   
    subject     = best_subj,
    model_name  = "cnnbilstm",
    process     = "raw",
    sensors     = ["LB"], 
    window_size = 200,
    stride      = 50,
    out_dir     = "../checkpoints/CNNBiLSTM/raw_only_back"
)

print(f"F1: {f1_score(y_true, y_pred, average='macro'):.4f}")
```

```python
best_subj_meta = load_trial(base_path, best_subj_trial)
best_subj_meta = best_subj_meta["metadata"] 
best_subj_cohort = best_subj_meta["pathologyKey"] 
best_subj_group = best_subj_meta["group"] 

worst_subj_meta = load_trial(base_path, worst_subj_trial)
worst_subj_meta = worst_subj_meta["metadata"] 
worst_subj_cohort = worst_subj_meta["pathologyKey"] 
worst_subj_group = worst_subj_meta["group"] 
```

```python
path = base_path + f"/{best_subj_group}/{best_subj_cohort}/{best_subj}/{best_subj_trial}"

plot_gait_detection(
    y_true, 
    best_subj_trial, 
    path, 
    title=f'Visualization of gait events for best predicted subject {best_subj} on trial 1 (GROUND TRUTH)', 
    sensor='LB', 
    signal_channel='Acc_Z', 
    process='raw')

plot_gait_detection(
    y_pred, 
    best_subj_trial, 
    path, 
    title=f'Visualization of gait events for best predicted subject {best_subj} on trial 1 (PREDICTED)', 
    sensor='LB', 
    signal_channel='Acc_Z', 
    process='raw')
```

```python
y_pred, y_true, signal = predict_trial(
    base_path   = base_path,
    trial_name  = worst_subj_trial,   
    subject     = worst_subj,
    model_name  = "cnnbilstm",
    process     = "raw",
    sensors     = ["LB"],
    window_size = 200,
    stride      = 50,
    out_dir     = "../checkpoints/CNNBiLSTM/raw_only_back"
)

print(f"F1: {f1_score(y_true, y_pred, average='macro'):.4f}")
```

```python
path = base_path + f"/{worst_subj_group}/{worst_subj_cohort}/{worst_subj}/{worst_subj_trial}"

plot_gait_detection(
    y_true, 
    worst_subj_trial, 
    path, 
    title=f'Visualization of gait events for worst predicted subject {worst_subj} on trial 1 (GROUND TRUTH)', 
    sensor='LB', 
    signal_channel='Acc_Z', 
    process='raw')

plot_gait_detection(
    y_pred, 
    worst_subj_trial, 
    path, 
    title=f'Visualization of gait events for worst predicted subject {worst_subj} on trial 1 (PREDICTED)', 
    sensor='LB', 
    signal_channel='Acc_Z', 
    process='raw')
```

It is clear that there is a big difference in the mean value of the best subject's trial and the worst one's. This might be the reason why this trial is predicted so poorly. In order to try to solve this we're going to apply a different normalization, instead of a global normalization (using just the trining data to avoid data leakage), we're going to normalize per window, that way discrepancies between the mean value for different subjects gets minimized.

After applying window normalization, the mean F1 score increased from 0.853 to 0.863. However, the worst subject still shows a very low F1 score (around 0.28).

Visual inspection of the predictions reveals that the model correctly detects the timing of gait events. When a gait event occurs, the prediction aligns closely with the ground truth. However, the model frequently confuses the left and right classes. In other words, left events are often predicted as right events and vice versa, while the no-event class is predicted correctly.

This suggests that the model is capable of detecting the presence of a step but struggles to determine which foot generated the event. This is a common issue that comes from the fact just the lower back sensor is being used. With just this sensor we successfully detect steps but we lack the laterality informaion that would help to correctly classify the most difficult examples.

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import sys
import os

sys.path.append(os.path.abspath(os.path.join('..')))

from src.load_data_gait import *
from src.plot_data_gait import *
from src.train_LSTMGait import train_model_cv
```

```python
base_path = '../datasets/GAIT_signals/dataset/data'
```

# Preprocessed data

```python
X, y, subjects = load_dataset_gait(base_path, process="preprocessed", window_size=200, stride=50)
X = X.astype(np.float32)
y = y.astype(np.int64)
```

```python
models_per_subject, test_subject_accuracies = train_model_cv(X, y, subjects, epochs=100, lr=0.0003, patience=10)
```

```python
f1_scores = np.vstack((pd.unique(subjects), test_subject_accuracies))
f1_scores = pd.DataFrame(f1_scores.T, columns=['patient', 'score'])
f1_scores.head()
```

```python
f1_scores.to_csv("../results/results_preprocessed_gait_lstm.csv", index=False)
```

```python
results = pd.read_csv("../results/results_preprocessed_gait_lstm.csv")
```

```python
print(f'Mean f1-score: {np.mean(results["score"])}')
print(f'Standard deviation: {np.std(results["score"])}')
```

```python
healthy = 'lightgreen'
ortho = 'lightcoral'
neuro = 'lightblue'

cohorts = {
    'HS': (("healthy", healthy), []), 
    'ACL': (("ortho", ortho), []), 
    'HOA': (("ortho", ortho), []), 
    'KOA': (("ortho", ortho), []), 
    'CIPN': (("neuro", neuro), []), 
    'CVA': (("neuro", neuro), []), 
    'PD': (("neuro", neuro), []), 
    'RIL': (("neuro", neuro), [])}



print('Mean f1-score per cohort:')

for cohort in cohorts.keys():
    for patient_name in results['patient']:
        if cohort in patient_name:
            score = results.loc[results['patient'] == patient_name, 'score'].values[0]
            cohorts[cohort][1].append(score)  


plt.figure(figsize=(12,8))

for cohort in cohorts.keys():
    plt.bar(cohort, np.mean(cohorts[cohort][1]), label=cohorts[cohort][0][0], color=cohorts[cohort][0][1])

plt.legend()


```

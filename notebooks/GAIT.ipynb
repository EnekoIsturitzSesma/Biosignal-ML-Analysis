```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import gc 
from sklearn.metrics import f1_score

import sys
import os

sys.path.append(os.path.abspath(os.path.join('..')))

from src.load_data_gait import *
from src.plot_data_gait import *
from src.train_LSTMGait import train_model_cv, predict_trial
```

```python
base_path = '../datasets/GAIT_signals/dataset/data'
```

# Preprocessed data

## Training

```python
X, y, subjects = load_dataset_gait(base_path, process="preprocessed", window_size=200, stride=50)
X = X.astype(np.float32)
y = y.astype(np.int64)
```

```python
models_per_subject, results_list = train_model_cv(X, y, subjects, epochs=100, lr=0.0003, patience=10, out_dir="../checkpoints/processed")
```

## Analysis

```python
f1_scores = np.vstack((np.unique(subjects), results_list))
f1_scores = pd.DataFrame(f1_scores.T, columns=['patient', 'score'])
f1_scores.head()
```

```python
f1_scores.to_csv("../results/results_preprocessed_gait_lstm.csv", index=False)
```

```python
results_processed = pd.read_csv("../results/results_preprocessed_gait_lstm.csv")
```

```python
print(f'Mean f1-score: {np.mean(results_processed["score"])}')
print(f'Standard deviation: {np.std(results_processed["score"])}')
```

```python
healthy = 'lightgreen'
ortho = 'lightcoral'
neuro = 'lightblue'

cohorts = {
    'HS': (("healthy", healthy), []),
    'ACL': (("ortho", ortho), []),
    'HOA': (("ortho", ortho), []),
    'KOA': (("ortho", ortho), []),
    'CIPN': (("neuro", neuro), []),
    'CVA': (("neuro", neuro), []),
    'PD': (("neuro", neuro), []),
    'RIL': (("neuro", neuro), [])}


for cohort in cohorts.keys():
    for patient_name in results_processed['patient']:
        if cohort in patient_name:
            score = results_processed.loc[results_processed['patient'] == patient_name, 'score'].values[0]
            cohorts[cohort][1].append(score)

mean = np.mean(results_processed["score"])

plt.figure(figsize=(12,8))

for cohort in cohorts.keys():
    bar = plt.bar(
        cohort,
        np.mean(cohorts[cohort][1]),
        yerr=np.std(cohorts[cohort][1]),
        capsize=5,
        label=cohorts[cohort][0][0],
        color=cohorts[cohort][0][1])

    plt.bar_label(bar, fmt='%.4f')

plt.axhline(y=mean, color='black', linestyle='--', label=f'Mean = {mean:.4f}')

plt.ylim(0.4, 1.05)

plt.legend()
plt.show()



```

```python
del X, y, subjects, models_per_subject
gc.collect()
```

# Prediction Visualization

```python
best_subj_idx = np.argmax(results_raw['score'])
best_subj = results_raw.iloc[best_subj_idx]['patient'] 
print('Subject with the best score:')
print(results_raw.iloc[best_subj_idx])
print('')

worst_subj_idx = np.argmin(results_raw['score'])
worst_subj = results_raw.iloc[worst_subj_idx]['patient'] 
print('Subject with the worst score:')
print(results_raw.iloc[worst_subj_idx])
print('')
```

```python
best_subj_trial = f"{best_subj}_1"  # Obtain first trial of the best classified subject 
worst_subj_trial = f"{worst_subj}_1"  # Obtain first trial of the worst classified subject 

print(best_subj_trial)
print(worst_subj_trial)
```

```python
y_pred, y_true, signal = predict_trial(
    base_path  = base_path,
    trial_name = best_subj_trial,   
    subject    = best_subj,
    process    = "raw",
    window_size = 200,
    stride      = 50,
    out_dir     = "../checkpoints/raw"
)

print(f"F1: {f1_score(y_true, y_pred, average='macro'):.4f}")
```

```python
path = base_path + f"/neuro/RIL/{best_subj}/{best_subj_trial}"

plot_gait_detection(
    y_true, 
    best_subj_trial, 
    path, 
    title=f'Visualization of gait events for best predicted subject {best_subj} on trial 1 (GROUND TRUTH)', 
    sensor='LB', 
    signal_channel='Acc_Z', 
    process='raw')

plot_gait_detection(
    y_pred, 
    best_subj_trial, 
    path, 
    title=f'Visualization of gait events for best predicted subject {best_subj} on trial 1 (PREDICTED)', 
    sensor='LB', 
    signal_channel='Acc_Z', 
    process='raw')
```

```python
y_pred, y_true, signal = predict_trial(
    base_path  = base_path,
    trial_name = worst_subj_trial,   
    subject    = worst_subj,
    process    = "raw",
    window_size = 200,
    stride      = 50,
    out_dir     = "../checkpoints/raw"
)

print(f"F1: {f1_score(y_true, y_pred, average='macro'):.4f}")
```

```python
path = base_path + f"/ortho/HOA/{worst_subj}/{worst_subj_trial}"

plot_gait_detection(
    y_true, 
    worst_subj_trial, 
    path, 
    title=f'Visualization of gait events for worst predicted subject {worst_subj} on trial 1 (GROUND TRUTH)', 
    sensor='LB', 
    signal_channel='Acc_Z', 
    process='raw')

plot_gait_detection(
    y_pred, 
    worst_subj_trial, 
    path, 
    title=f'Visualization of gait events for worst predicted subject {worst_subj} on trial 1 (PREDICTED)', 
    sensor='LB', 
    signal_channel='Acc_Z', 
    process='raw')
```
